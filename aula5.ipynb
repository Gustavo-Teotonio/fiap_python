Ex.6 - Encontrando a soma dos números em uma lista usando while 


```python
lista = [23, 45, 12, 10, 25] 
resultado = 0
i = 0

while(i < len(lista)):
    resultado += lista[i]
    i += 1

print(f"Resultado: {resultado}")
```

Ex.8 - Crie um programa que solicite 5 números inteiros e 
decremente os valores inseridos pelo usuário.

```python
lista = [] 
resultado = 0
i = 0

while(i < len(lista)):
    lista.append(int(input("Digite um número: ")))
    i += 1
    
resultado = 0

while(len(lista) > 0):
    resultado -= lista.pop()

print(f"Resultado: {resultado}")
```

Ex.9 - Crie um programa para gerar uma taboada, a partir 
de duas variáveis (numero e limite).

```python
numero = int(input("digite o número: "))
limite = int(input("digite o limite: "))
indice = 0


while(indice <= limite):
    print(f"{numero} x {indice} = {numero * indice}")
    indice += 1
```

Ex.10 - A partir da var word = 'engenharia de software'. Se a 
indice da word == 'e' or 'o', continue. 

```python
word = 'engenharia de software'
lista_caracteres = word.split()
i = 0

while(i < len(lista_caracteres)):
    if(lista_caracteres[i] == 'e' or lista_caracteres[i] == 'o'):
        i += 1
        continue
    print(lista_caracteres[i])
    i += 1

```

Ex.11 - Cadastro de alunos: crie um algoritmo que solicite 3 nomes e 3 idades.
Deixe a informação em uma lista no formato nome - idade.


```python
lista_nome_idade = []
i = 0

while(i < 3):
    nome = input("Digite o nome: ")
    idade = int(input("Digite a idade: "))
    lista_nome_idade.append((nome, idade))
    i += 1

print(f"Nomes e idades digitados: {lista_nome_idade}")
```


Ex.12 - Crie uma estrutura que solicite itens de compra de forma que só seja possível sair da estrutura de repetição, caso o usuário digite sair. Coloque os itens em uma lista.

```python
lista_compra = []

while True:
    item = input("Digite o item da compra (ou sair para finalizar): ")

    if(item.lower() == 'sair'):
        break

    lista_compra.append(item)
    
print(f'Lista de compras: {lista_compra}')
```

Ex.13 - Crie um programa que solicite 5 números inteiros, armazene em um lista o seu valor ao quadrado. [numero, numero**2]

```python
i = 0
lista_valores = []

while (i < 5):
    valor = int(input("Digite um número: "))
    lista_valores.append((valor, valor**2))
    i += 1

print(f"Valores totais: {lista_valores}")
```

Ex.14 - Crie um programa que solicite nome, curso e idade de 3 pessoas, armazene cada pessoa como uma lista dentro de uma lista_completa.


```python
i = 0
lista_completa = []

while(i < 3):
    lista_cadastro = []
    nome = input("Digite o nome: ")
    curso = input("Digite o nome do curso: ")
    idade = int(input("Digite a idade: "))
    lista_cadastro.append(nome)
    lista_cadastro.append(curso)
    lista_cadastro.append(idade)
    lista_completa.append(lista_cadastro)
    i += 1

print(lista_completa)
```


Ex.15 - Crie um programa que solicite números para o usuário, pare quando digitar 0, armazene os números em uma lista. Mostre a lista completa e a soma dos valores.

```python
lista_valores = []

while True:
    numero = int(input("Digite um número (0 para sair): "))
    if(numero == 0):
        break
    lista_valores.append(numero)

i = 0
soma = 0

while(i < len(lista_valores)):
    soma += lista_valores[i]
    i += 1  


print(f"""Soma dos números digitados: {soma}
Lista completa: {lista_valores}
""")    
```
