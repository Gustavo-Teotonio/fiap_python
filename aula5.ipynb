Ex.6 - Encontrando a soma dos números em uma lista usando while 


```python
lista = [23, 45, 12, 10, 25] 
resultado = 0
i = 0

while(i < len(lista)):
    resultado += lista[i]
    i += 1

print(f"Resultado: {resultado}")
```


Ex.7 - Crie um programa que solicite 5 nomes e insira em 
uma lista.

```python
lista_nome = []

while(len(lista_nome) < 5):
    nome = input("Digite um nome: ")
    lista_nome.append(nome)

print(f"Nomes digitados: {lista_nome}")
```

Ex.8 - Crie um programa que solicite 5 números inteiros e 
decremente os valores inseridos pelo usuário.

```python
lista = [] 
resultado = 0
i = 0

while(i < len(lista)):
    lista.append(int(input("Digite um número: ")))
    i += 1
    
resultado = 0

while(len(lista) > 0):
    resultado -= lista.pop()

print(f"Resultado: {resultado}")
```

Ex.9 - Crie um programa para gerar uma taboada, a partir 
de duas variáveis (numero e limite).

```python
numero = int(input("digite o número: "))
limite = int(input("digite o limite: "))
indice = 0


while(indice <= limite):
    print(f"{numero} x {indice} = {numero * indice}")
    indice += 1
```

Ex.10 - A partir da var word = 'engenharia de software'. Se a 
indice da word == 'e' or 'o', continue. 

```python
word = 'engenharia de software'
lista_caracteres = word.split()
i = 0

while(i < len(lista_caracteres)):
    if(lista_caracteres[i] == 'e' or lista_caracteres[i] == 'o'):
        i += 1
        continue
    print(lista_caracteres[i])
    i += 1

```
