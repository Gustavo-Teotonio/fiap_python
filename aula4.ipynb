## Ex.5 - Soma a média de valores digitados pelo usuário. Crie um algoritmo para solicitar 5 valores não inteiros, insira em uma lista. após isso, crie uma estrutura para somar e retornar a média dos elementos desta lista

```python
lista_valores = []

for i in range(5):
    num = float(input('Digite um número: '))
    lista_valores.append(num)

valor = 0

for i in lista_valores:
    valor += i

print(f'Média: {valor / len(lista_valores)}')
```

## Ex.6 - Crie uma estrutura para solicitar 10 número para o usuário. Após isso crie duas listas (par e ímpar) e insira os valores pares na lista par, valores ímpares na lista ímpar. Print a lista completa, lista par e lista ímpar.

```python
lista_impar = []
lista_par = []

for i in range(10):
    num = float(input('Digite um número: '))
    if num % 2 == 0:
        lista_par.append(num)
    else:
        lista_impar.append(num)

print(f'Números pares: {lista_par}')
print(f'Números ímpares: {lista_impar}')
```

## Ex.7 - Crie um algoritmo para cadastro de alunos, onde deve ser solicitado nome e idade 5 vezes, porém a informação deve aparecer com lista dentro de lista, ou seja, [nome1, idade1], [nome2, idade2] e assim por diante. Print a lista final

```python
lista_pessoas = []
lista_cadastro = []

for i in range(3):
    nome = input('Digite o nome da pessoa: ')
    idade = int(input('Digite a idade da pessoa: '))
    lista_pessoas.append(nome)
    lista_pessoas.append(idade)
    lista_cadastro.append(lista_pessoas)
    lista_pessoas = []

print(lista_cadastro)
```

## Ex. 8 - Vamos fazer um algoritmo para cadastro com média. Solicite 3 informações: nome, nota1 e nota 2 três vezes. Apos isso, calcule a media de cada aluno e retorne uma lista contendo nome, nota1, nota 2, media de cada aluno. Estrutura: lista = [[nome, nota1, nota2, media], [nome, nota1, nota2, media]]

```python
lista_notas = []
lista_cadastro = []

for i in range(3):
    nome = input('Digite o nome do aluno: ')
    nota1 = float(input('Digite a primeira nota: '))
    nota2 = float(input('Digite a segunda nota: '))
    media = (nota1 + nota2) / 2
    lista_cadastro.append(nome)
    lista_cadastro.append(nota1)
    lista_cadastro.append(nota2)
    lista_cadastro.append(media)
    lista_notas.append(lista_cadastro)
    lista_cadastro = []

print(lista_notas)
```

# Ex.9 - Crie um algoritmo para simular um sistema de compras, onde devera ser solicitado nome 100x. Após isso, crie uma estrutura para quebrar a solicitacao de nomes. apos isso crie uma estrutura para solicitar preco, com isso, deixe o respectivo nome e preco juntos. Finalmente, mostre o valor total dos produtos         

```python
lista_compra = []   
lista_nome_itens = []
soma = 0

for i in range(100):
    nome = input('Digite o nome do produto: \n[Q] para sair')
    if nome.upper() == 'Q':
        break
    lista_nome_itens.append(nome)

for i in range(len(lista_nome_itens)):
    preco = float(input(f'Digite o preço do produto {lista_nome_itens[i]}: '))
    lista_compra.append((lista_nome_itens[i], preco))
    soma += preco
    
print(f'Lista de compras: {lista_compra}')
print(f'Total da compra: {soma}')


```

# Ex.10 - Crie um algoritmo para solicitar nome e nota para 4 alunos. Insira em uma estrutura com lista dentro de lista (nome, nota). Após isso, crie uma estrutura para mostrar os alunos aprovados com nota maior ou igual a 6, mostrando o respectivo nome e nota. Crie também uma estrutura para mostrar os alunos com nota menor do que 6 e mostre o respectivo nome e nota

```python
lista_aluno_aprovado = []
lista_aluno_reprovado = []

for i in range(4):
    nome = input("Digite o nome do aluno: ")
    nota = float(input("Digite a nota do aluno: "))

    if(nota >= 6):
        lista_aluno_aprovado.append((nome, nota))
    else:
        lista_aluno_reprovado.append((nome, nota))

print(f"Alunos aprovados: {lista_aluno_aprovado}")
print(f"Alunos reprovados: {lista_aluno_reprovado}")
```

# Ex.11 - Crie uma algoritmo para simular sistema com menu. O usuário pode cadastrar, listar ou sair da estrutura. Se a opção for 1, solicite o nome, idade, cidade e deixe estas informações em uma estrutura de lista dentro de lista. Se opção 2, print o respectivo nome, idade e cidade da pessoa. Se opção for 3, crie uma estrutura para sair do código

```python
lista_cadastro = []

for i in range(100):
    opcao_menu = int(input("""
    [1] - Cadastrar aluno
    [2] - Printar aluno
    [3] - Sair
    """))

    if(opcao_menu == 1):
        nome = input("Digite o nome do aluno: ")
        idade = int(input("Digite a idade do aluno: "))
        cidade = input("Digite a cidade do aluno: ")
        lista_cadastro.append((nome, idade, cidade))
    elif(opcao_menu == 2):
        for obj in lista_cadastro:
            print(f"Nome: {obj[0]}, Idade: {obj[1]}, Cidade: {obj[2]}")
    elif(opcao_menu == 3):
        print("Até logo!")
        break
    else:
        print("Opção inválida, favor escolha uma opção válida.")
```

# Ex.12 - Criando uma matriz (3x3) usando estrutura de repeticao

```python
lista_um = []

for i in range(3):
    lista_dois = []
    for k in range(3):
        lista_dois.append(k)
    lista_um.append(lista_dois)

print(lista_um)
```
