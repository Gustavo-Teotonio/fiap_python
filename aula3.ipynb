#1. Sistema de login com nivel de acesso. solicite o usuario e senha. Se o usuário é igual a admin, crie uma estrutura de condição aninhada para solicitar senha e se a mesma for '1234', mostre que o usuário terá acesso total. Caso o usuário insira senha incorreta, mostre senha incorreta. Caso usuário insira usuário incorreto, mostre usuário incorreto

```python
usuario = input('Qual o usuario')

if(usuario == 'admin'):
    senha = input('Qual a senha: ')
    if(senha == '1234'):
        print('Login feito com sucesso, você possui acesso total')
    else:
        print('Senha incorreta')
else:
    print('Usuário inválido')
        
```

#2. Classifiação de idade. Solicite idade, se a idade for maior ou igual a 18, crie uma estrutura de condição aninhada para verificar se idade é maior ou igual a 60, se for., mostre que é idoso, senão mostre que é adulto. Se idade for maior ou igual a 12, adolescente, caso contrário, criança

```python
idade = int(input('Qual a idade'))

if(idade >= 18):
    if(idade >= 60):
        print('Idoso')
    else:
        print('Adulto')
elif(idade >= 12):
    print('Adolescente')
else:
    print('Criança')
```

#3. Aprovação com distinção. Solicite nota, se a nota for maior ou igual a 6, crie uma condição aninhada para verificar se nota é maior ou igual a 9, se for aprovado com excelência. Se a nota não for maior ou igual a 9, aprovado. Caso contrário, reprovado

```python
nota = float(input('Qual a nota'))

if(nota >= 6):
    if(nota >= 9):
        print('Aprovado com excelência')
    else:
        print('Aprovado')
else:
    print('Reprovado')
```

#4. Verificação de número. Solicite número e verifique se é maior do que zero, se for, crie uma estrutura de condição aninhada para verificar se este número é par, se for, print positivo e par. Se não for, Positivo e impar. Se número for igual a zero, print zero, caso contrário, printe negativo

```python
numero = int(input('Número: '))

if(numero > 0):
    if(numero % 2 == 0):
        print('Par e positivo')
    else:
        print('Impar e positivo')
elif(numero < 0):
    if(numero % 2 == 0):
        print('Par e negativo')
    else:
        print('Impar e negativo')
else:
    print('0')

```

#5. Sistema de desconto. solicite valor e se a pessoa é vip ou não. Se valor maior ou igual a 200, crie uma estrutura de condição aninhada para verificar se a pessoa é vip, se for, ofereça 20% de desconto sobre o valor e mostre o valor a ser descontado e o valor final, considerando o desconto. Se não for vip, ofereça um desconto de 10%

```python
valor = int(input('Qual o valor'))

if(valor >= 200):
    vip = input('Você é vip [SIM/NAO]').upper()
    if(vip == 'SIM'):
        valor_atual = valor * 0.8
        print(f'Valor descontado: R${valor - valor_atual:.2f}\nValor a ser pago: R${valor_atual:.2f}') 
    else:
        valor_atual = valor * 0.9
        print(f'Valor descontado: R${valor - valor_atual:.2f}\nValor a ser pago: R${valor_atual:.2f}') 
else:
    print('Ta duro dorme pai')

```

# 6. Crie um algoritmo para perguntar para o usuário qual o dia da semana, caso seja sábado, escreva dia de festa, caso seja domingo, pergunte sobre a condição fisica do usuário; se estiver com dores de cabeça, print recuperando, então precisa descansar. Caso contrário, apenas descanse. Caso não seja sábado ou domingo, mostre trabalhando, trabalhando e trabalhando

```python
dia_semana = input('Qual o dia da semana ? [Sab/Dom]')

if(dia_semana == 'Sab'):
    print('Dia de festa')
elif(dia_semana == 'Dom'):
    condicao_fisica = input('Qual a sua condição fisica [Dor de cabeça/De boa]')
    if(condicao_fisica == 'Dor de cabeça'):
        print('Recuperando')
    else:
        print('Descanse')
else:
    print('Workando hard!')
```

# 7. Crie um algoritmo para solicitar a escolha do menu do café da manhã
1. Egs
2. Pancakes
3. Waffles
4. Frutas

Se for eggs, pergunte qual o tipo de acompanhamento
Se for frutas, pergunte qual o tipo de fruta e acompanhamento

Para cada caso, traga uma informação com print para o usuário

```python
escolha = int(input(""" Qual a sua escolha?
                        [1] - Eggs
                        [2] - Pancakes
                        [3] - Waffles
                        [4] - Frutas
"""))

if(escolha == 1):
    acompanhamento = int(input("""Qual acompanhamento você deseja?
                                  [1] - Bacon
                                  [2] - Queijo
                                  [0] - Não quero
                               """))
    if(acompanhamento == 1):
        print('Seu café da manhã será composto de ovos com bacon')
    elif(acompanhamento == 2):
        print('Seu café da manhã será composto de ovos com queijo')
    elif(acompanhamento == 0):
        print('Seu café da manhã será composto apenas de ovos')
    else:
        print('Café da manhã cancelado')
elif(escolha == 2):
    print('Seu café da manhã será composto apenas de panquecas')

elif(escolha == 3):
    print('Seu café da manhã será composto apenas de Waffles')

elif(escolha == 4): 
    tipo_fruta = int(input("""Qual tipo de fruta você deseja:
                              [1] - Banana
                              [2] - Maça
                           """))
    
    acompanhamento = int(input("""Qual acompanhamento você deseja?
                                  [1] - Granola
                                  [2] - Leite condensado
                               """))
    if(tipo_fruta == 1 and acompanhamento == 1):
        print('Seu café da manhã será composto de Banana com granola')
    elif(tipo_fruta == 1 and acompanhamento == 2):
        print('Seu café da manhã será composto de Banana com leite condensado')
    elif(tipo_fruta == 2 and acompanhamento == 1):
        print('Seu café da manhã será composto de Maça com granola')
    elif(tipo_fruta == 2 and acompanhamento == 2):
        print('Seu café da manhã será composto de Maça com leite condensado')
    else:
        print('Café da manhã cancelado')
```
